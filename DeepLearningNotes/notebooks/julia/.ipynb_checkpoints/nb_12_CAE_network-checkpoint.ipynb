## Convolutional Autoencoder for MNIST data

### Loading the data

First we load the required packages:

```julia
using Flux, Flux.Data.MNIST
using Flux: @epochs, onehotbatch, mse, throttle
using Base.Iterators: partition
```

```julia
# Encode MNIST images as compressed vectors that can later be decoded back into
# images.

imgs = MNIST.images();
labs = MNIST.labels();
# Partition into batches of size 1000
data = [float(hcat(vec.(imgs)...)) for imgs in partition(imgs, 1000)];
```

```julia
N = 2 # Size of the encoding
```

```julia
################################# Define Model #################################

encoder = Dense(28^2, N, leakyrelu)
decoder = Dense(N, 28^2, leakyrelu)

m = Chain(encoder, decoder)
```

```julia
# Size of the encoding for each layer
N1 = 512 
N2 = 64
N3 = 16

#Define the layers of the neural network, with the inner layer having dimension 2 and a sigmoidal activation function
encoder1 = Dense(28^2, N1, leakyrelu) 
encoder2 = Dense(N1, N2, leakyrelu) 
encoder3 = Dense(N2, N3, leakyrelu) 
encoder4 = Dense(N3, 2, sigmoid) 
decoder4 = Dense(2, N3, leakyrelu) 
decoder3 = Dense(N3, N2, leakyrelu) 
decoder2 = Dense(N2, N1, leakyrelu)
decoder1 = Dense(N1, 28^2, leakyrelu)

#Define the model and loss function
encoder = Chain(encoder1, encoder2, encoder3, encoder4)
decoder = Chain( decoder4, decoder3, decoder2, decoder1)
# m = Chain(encoder1, encoder2, encoder3, encoder4, decoder4, decoder3, decoder2, decoder1)
m = Chain(encoder,decoder)
loss(x) = mse(m(x), x)

evalcb = throttle(() -> @show(loss(data[1])), 5)
opt = ADAM();
```

```julia
################################# Learn Parameters ##############################

@epochs 10 Flux.train!(loss, params(m), zip(data), opt, cb = evalcb)
```

```julia
# Sample output

using Images

img(x::Vector) = Gray.(reshape(clamp.(x, 0, 1), 28, 28))

function sample()
  # 20 random digits
  before = [imgs[i] for i in rand(1:length(imgs), 20)]
  # Before and after images
  after = img.(map(x -> m(float(vec(x))).data, before))
  # Stack them all together
  hcat(vcat.(before, after)...)
end

save("sample_sae.png", sample())
```

```julia
using Plots
cols = [labs[i] for i in 1:2000]
enc = [ encoder(float(vec(imgs[i]))).data  for i in 1:2000 ] #rand(1:length(imgs), 2000)]
enc1 = first.(enc[cols .== 0])
enc2 = last.(enc[cols .== 0])
p = scatter(enc1,enc2,label=0)
for lab = 1:9
    enc1 = first.(enc[cols .== lab])
    enc2 = last.(enc[cols .== lab])
    scatter!(enc1,enc2,label=lab)
end
p
```
