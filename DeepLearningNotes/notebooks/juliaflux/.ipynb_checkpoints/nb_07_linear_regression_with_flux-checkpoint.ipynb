# Linear Regression with Julia and Flux

### One dimensional linear regression

```julia
# generate some test data
using Plots
regX = rand(100)
regY = 50 .+ 100 * regX + 2 * randn(100)
scatter(regX, regY, fmt = :png, legend=:bottomright, label="data")
```

```julia
regX[1]
```

```julia
using Flux
using Flux: throttle

data = zip(regX, regY)  # create tuples
model = Dense(1, 1, identity) # define the model
loss(x, y) = Flux.mse(model([x]), y) # use mean square error for loss

# define a call back ftn that prints the loss over the whole data set
evalcb = () -> @show(sum([loss(i[1],i[2]) for i in data]))

# use Stocastic Gradient Descent to optimise the model weights
# opt = SGD(Flux.params(modelReg), 0.1)
opt = Descent(0.1)

# train the model on 10 epochs
for i = 1:10 Flux.train!(loss, params(model), data, opt, cb = throttle(evalcb, 1)) end
# model([regX[1]])
```

```julia
# display the linear fit by evaluating the model
# on evenly spread data points
predX = collect(range(0, 1,length=100))
predY = model(predX').data
scatter(regX, regY, fmt = :png, legend=:bottomright, label="data")
scatter!(predX, predY', label="fit")
```

```julia
# print out the two model weights and 
# compare to the lin alg solution
println(Tracker.data.(Flux.params(model)))
```

### Two dimensional linear regression

```julia
# generate data over two dimensions
using Plots
regX = 2 .* (rand(500) .- 0.5)
regY = 2 .* (rand(500) .- 0.5)

# Z = f(X,Y)
# regZ = 10 .+ ( ( regX.^2 + regY.^2) .+ 0.1 * (randn(500) .- 0.5) )
regZ = 10 .+ 5 .* regX + 12 .* regY .+ 0.2 * (randn(500) .- 0.5) 
scatter(regX, regY, regZ,  fmt = :png, legend=:bottomright, label="data")
```

```julia
using Flux
using Flux: throttle

data2D = [([regX[i],regY[i]],regZ[i]) for i in 1:length(regX)]
model = Chain(Dense(2, 1, identity))
loss(x, y) = Flux.mse(model(x), y) 
evalcb = () -> @show(sum([loss(i[1],i[2]) for i in data2D]))
opt = ADAM(0.1)
for i=1:50 Flux.train!(loss, params(model), data2D, opt, cb = throttle(evalcb, 1000)) end
```

```julia
println(Flux.params(model))
```

```julia
predX = collect(range(-1, 1,length=10))
predY = collect(range(-1, 1,length=10))
predXY = [ [ predX[i], predY[j] ] for i in 1:10 for j in 1:10 ]
predZ = Tracker.data.( model.(predXY) )
predD = zeros(0)
for i in 1:length(predZ)
    append!(predD , [ predXY[i][1], predXY[i][2], predZ[i][1] ])  
end
predD = reshape(predD,3,length(predZ))'
# Tracker.data( modelReg.([[-1,0],[1,0]]) )
# predD[:,1]
scatter(regX, regY, regZ, fmt = :png, legend=:bottomright, label="data")
scatter!( predD[:,1], predD[:,2], predD[:,3], label="fit" )
```

### Two dimensional Quadratic regression

```julia
using Plots
regX = 2 .* (rand(1000) .- 0.5)
regY = 2 .* (rand(1000) .- 0.5)
regZ = 10 .+ ( ( regX.^2 + regY.^2) .+ 0.1 * (randn(1000) .- 0.5) )
scatter(regX, regY, regZ, fmt = :png, legend=:bottomright, label="data")
```

```julia
using Flux

dataNL = [([regX[i],regY[i]],regZ[i]) for i in 1:length(regX)]
model = Chain(Dense(2, 1, identity))
loss(x, y) = Flux.mse(model(x), y) 
evalcb = () -> @show(sum([loss(i[1],i[2]) for i in dataNL]))
opt = ADAM(0.1)
for i=1:30 Flux.train!(loss,params(model), dataNL, opt, cb = throttle(evalcb, 10)) end
```

```julia
println(Flux.params(model))
```

```julia
predX = collect(range(-1, 1,length=10))
predY = collect(range(-1, 1,length=10))
predXY = [ [ predX[i], predY[j] ] for i in 1:10 for j in 1:10 ]
predZ = Tracker.data.( model.(predXY) )
predD = zeros(0)
for i in 1:length(predZ)
    append!(predD , [ predXY[i][1], predXY[i][2], predZ[i][1] ])  
end
predD = reshape(predD,3,length(predZ))'
# Tracker.data( modelReg.([[-1,0],[1,0]]) )
# predD[:,1]
scatter(regX, regY, regZ, fmt = :png, legend=:bottomright, label="data")
scatter!( predD[:,1], predD[:,2], predD[:,3], label="fit" )
```
